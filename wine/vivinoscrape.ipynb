```python
def vivino_scrape(wine_name: str):
    url= "http://www.vivino.com/search/wines?q="+ wine_name.replace(" ","+")
    vivino_headers = {"User-Agent": "Mozilla/5.0 (Windows NT 6.1; Win64; x64; rv:72.0) Gecko/20100101 Firefox/72.0", "Accept-Language": "en-US,en;q=0.5"}
    
    page = requests.get(url,headers = vivino_headers)
    
    vivino_html = page.text
    vivino_soup = BeautifulSoup(vivino_html)
    vivino_rating = vivino_soup.find("div",{"class":"text-inline-block light average__number"}).text.strip()
    vivino_name=vivino_soup.find("span",{"class":"bold"}).text.strip()
    vivino_price=vivino_soup.find("span",{"class":"wine-price-value"}).text.strip()
    vivino_numberofreviews=vivino_soup.find("p",{"class":"text-micro"}).text.strip().replace("ratings"," ")
    
    return {"name":vivino_name, "price": vivino_price,"rating":vivino_rating ,"numberofreviews":vivino_numberofreviews}
```

```python
data=dict()
for w in wine_names_text:
    data[w]=vivino_scrape(w)
```

```python
import pandas as pd
df=pd.DataFrame.from_dict(data,orient="index")
df
```

```python
df=df.reset_index(drop=True)
```

```python
df.to_excel("carrefourvivino.xlsxvino.xlsx",index=False)
```
